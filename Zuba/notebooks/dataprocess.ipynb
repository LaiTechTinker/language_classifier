```python
#now let's remove all the neutral and negative labels from all the language datasets so that we can only get only positive data
files_path=[
"../data/converted_dataset/hau_test.csv",
"../data/converted_dataset/ibo_test.csv",
"../data/converted_dataset/pcm_test.csv",
"../data/converted_dataset/yor_test.csv"
]
def remove_neutral_negative_labels(file_path):
  for file in files_path:
    df=pd.read_csv(file)
    df=df[~df["label"].isin(["negative","neutral"])]
    df.to_csv(file,index=False)
remove_neutral_negative_labels(files_path)
```

```python
#now let's convert each label to numerical values base on their file name
os.makedirs("../data/new_dataset",exist_ok=True)
new_file_path=[
"../data/new_dataset/hau_test.csv",
"../data/new_dataset/ibo_test.csv",
"../data/new_dataset/pcm_test.csv",
"../data/new_dataset/yor_test.csv"
]
old_file_path=[
  "../data/converted_dataset/hau_test.csv",
"../data/converted_dataset/ibo_test.csv",
"../data/converted_dataset/pcm_test.csv",
"../data/converted_dataset/yor_test.csv"
]

def label_encoding(old_file_path,new_file_path):
  count=0
  for old_file,new_file in zip(old_file_path,new_file_path):
    df=pd.read_csv(old_file)
    df["label"]=df["label"].replace({"positive":count})
    df.to_csv(new_file,index=False)
    count+=1
label_encoding(old_file_path,new_file_path)
    
```

```python
#good job so far now let's convert the dataset into one large csv file
import pandas as pd

def combine_datasets(files_path, output_file):
    dfs = []  # store all dataframes here

    for file in files_path:
        df = pd.read_csv(file)
        dfs.append(df)

    # Combine into one big dataframe
    combined = pd.concat(dfs, ignore_index=True)

    # Save to CSV
    combined.to_csv(output_file, index=False)


combine_datasets(new_file_path, "../data/new_dataset/combined_data.csv")
```
